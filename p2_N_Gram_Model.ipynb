Practical - 2 : Implement N-Gram Language Model.

```python
# Import libraries
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# Download required NLTK data
nltk.download('punkt')
```

```python
# Sample text
text = "Natural Language Processing enables machines to understand human language."
```

```python
# Tokenize text
tokens = word_tokenize(text.lower())
```

```python
# Generate N-Grams (Unigram, Bigram, Trigram)
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("ðŸ”¹ Unigrams:\n", unigrams)
print("\nðŸ”¹ Bigrams:\n", bigrams)
print("\nðŸ”¹ Trigrams:\n", trigrams)
```

```python
# Frequency Distribution
def plot_ngram_freq(ngram_list, title):
    freq_dist = Counter(ngram_list)
    common = freq_dist.most_common(10)
    labels, values = zip(*common)

    # Convert tuple ngrams to string for labels
    labels = [' '.join(gram) for gram in labels]

    plt.figure(figsize=(8, 3))
    plt.bar(labels, values, color='skyblue')
    plt.xticks(rotation=45)
    plt.title(title)
    plt.ylabel('Frequency')
    plt.show()
    
# Plot N-Gram Frequency
plot_ngram_freq(unigrams, "Top 10 Unigrams")
plot_ngram_freq(bigrams, "Top 10 Bigrams")
plot_ngram_freq(trigrams, "Top 10 Trigrams")
    
```
